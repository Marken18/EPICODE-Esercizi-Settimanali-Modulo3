```python
#esercizio 1) round e clip

import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")



connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
query = """SELECT * 
FROM dimproduct"""
df = pd.read_sql(query,db_engine)
```

```python
df["DealerPrice_round"]=df["DealerPrice"].round(2)
df["DealerPrice_round"]
```

```python
df['DealerPrice_clipped'] = df['DealerPrice'].clip(lower=0, upper=1000)
df["DealerPrice_clipped"]
```

```python
#Esercizio 4) FACEBOOK
fb = pd.read_csv("datasets/beginner_datasets/facebook.csv")
fb.head()
```

```python
fb["status_published"] = pd.to_datetime(fb["status_published"])
fb.head()
```

```python
#Estrai informazioni specifiche dalle date e crea le nuove colonne
fb["status_published"]
fb['Anno'] = fb["status_published"].dt.year
fb['mese'] = fb["status_published"].dt.month
fb['giorno'] = fb["status_published"].dt.day
fb['giorno_della_settimana'] = fb["status_published"].dt.dayofweek  # 0=lunedì, 6=domenica
fb['giorno_dell_anno'] = fb["status_published"].dt.dayofyear

fb.head()
```

```python
post_2012 = fb[fb["Anno"] == 2012]
post_2012
```

```python
post_maggio_2018 = fb[(fb["Anno"] == 2018) & (fb['mese'] == 5)]
post_maggio_2018
```

```python
# Confrontiamo il numero di post pubblicati nei weekend rispetto al resto della settimana
# Weekend: sabato (5) e domenica (6)
weekend_posts = fb[fb['giorno_della_settimana'].isin([5,6])].shape[0]
weekend_posts
```

```python
# Resto della settimana
weekday_posts = fb[~fb['giorno_della_settimana'].isin([5,6])].shape[0]
weekday_posts	
```

```python
primi_ultimi_per_anno = fb.groupby('Anno')['status_published'].agg(['min', 'max']).reset_index()
primi_ultimi_per_anno.columns = ['Anno', 'primo_post', 'ultimo_post']
primi_ultimi_per_anno 
```

```python
tipi_post = fb['status_type'].value_counts()
numero_tipi = fb['status_type'].nunique()
print(tipi_post,numero_tipi)
```
