```python
#Esercizio 1) attività quadrato longitudine 9-10 e latitudine 45-46? Metodo BETWEEN
import pandas as pd
df = mappa
filter_longitudine = df["Longitudine"].between(9, 10, inclusive="both")


filter_latitudine = df["Latitudine"].between(45, 46, inclusive="both")


attivita_nel_quadrato = df[filter_longitudine & filter_latitudine]


numero_attivita = len(attivita_nel_quadrato)

print(f"Numero di attività nel quadrato di longitudine 9-10 e latitudine 45-46: {numero_attivita}")
```

```python
media_pH_per_colore = wine.groupby("type")["pH"].mean()
print("Media di pH per colore:")
print(media_pH_per_colore)
```

```python
media_solfati_per_qualità = wine.groupby("quality")["sulphates"].mean()
print("Media di solfati per qualità:")
print(media_solfati_per_qualità)
```

```python
media_solfati_per_colore = wine.groupby("type")["sulphates"].mean()
print("Media di solfati per colore:")
print(media_solfati_per_colore)
```

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

dotenv.load_dotenv(dotenv_path="cred.env", override=True)

username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")



connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

db_engine = sqlalchemy.create_engine(connection_string)
```

```python
#Esercizio su ADVENTUREWORKS
query = "SELECT * FROM dimemployee"
dimemployee = pd.read_sql(query,db_engine)
dimemployee
```

```python
import pandas as pd 
query = "SELECT * FROM dimemployeesalesterritory"
dimemployeesalesterritory = pd.read_sql(query,db_engine)
dimemployeesalesterritory
```

```python
df_join = pd.merge(dimemployee, dimemployeesalesterritory, on='EmployeeKey', how='inner')
print(f"Numero di record nel DataFrame risultante: {len(df_join)}")
```

```python
query = "SELECT * FROM dimsalesterritory"
dimsalesterritory = pd.read_sql(query,db_engine)
dimsalesterritory
```

```python
df_final = pd.merge(df_join, dimsalesterritory, on='SalesTerritoryKey', how='left')
```

```python
counts = df_final.groupby(['SalesTerritoryCountry', 'SalesTerritoryRegion']).size().reset_index(name='Number_of_Employees')
print(counts)
```

```python
avg_base_rate = df_final.groupby('SalesTerritoryCountry')['BaseRate'].mean()
print(avg_base_rate)
```
